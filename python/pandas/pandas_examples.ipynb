### Apply lambda function by row (axis=1)

```python
def bkt_sales(sales):
    if sales < 2000:
        bkt = 0
    elif (sales >= 2000) & (sales < 3500):
        bkt = 1
    else:
        bkt = 2
    return bkt
```

```python
df['bkt_SALES'] = df.apply(lambda x: bkt_sales(x['SALES']), axis=1)
```

```python
df[['SALES', 'bkt_SALES']]
```

### Merge Dataframes

```python
list_columns = list(df.columns)
list_cols1 = list_columns[:10]
list_cols2 = list_columns[10:]
```

```python
list_cols1
```

```python
list_cols2
```

```python
df1 = df.loc[:, list_cols1].copy()
df2 = df.loc[:, list_cols2].copy()
```

```python
df_merged = pd.merge(df1, df2, left_index=True, right_index=True, how='inner')
```
